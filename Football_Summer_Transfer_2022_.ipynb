# Transfrom Data


*  Convert dtype 'age' to int
*  Convert dtype 'cost' to float
*  Convert dtype 'date_of_transfer' to datetime
*  Delete non-numerical data in 'cost'
*  Replace string to related numerical data in 'cost'
*  Find Missing Data










```python
df['date_of_transfer'] = pd.to_datetime(df['date_of_transfer'], format='%Y-%m-%d') #converting 'date_of_transfer' from object into datetime
```

```python
df['age'] = df['age'].astype(int) #converting 'age' from object into int
```

```python
df=df[df["cost"].str.contains("Loan fee")==False]
df=df[df["cost"].str.contains("-")==False]
df=df[df["cost"].str.contains("\?")==False]
df=df[df["cost"].str.contains("loan transfer")==False]
df=df[df["cost"].str.contains("draft")==False]

df["cost"]=df["cost"].str.replace("free transfer","0")
df["cost"]=df["cost"].str.replace("Th.","000")
df["cost"]=df["cost"].str.replace("m","0000")
df["cost"]=df["cost"].str.replace(".","")
df["cost"]=df["cost"].str.replace("€","")

df["cost"]=df["cost"].astype(int)
```

```python
df.head(20)
```

```python
#Find missing Data
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis = 1, keys=['Total','Percent'])
missing_data.head(20)
```

```python
df['position'].fillna('Unknown', inplace=True)
```

```python
# Handling missing values 
for col_name in list(['league_new_club','country_new_club','league_origin_club','country_origin_club']):
     df[col_name].fillna('Free Transfer', inplace=True)
```

```python
df.head()
```

```python
# handling missing values Age fill with mean
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
#Find missing Data
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis = 1, keys=['Total','Percent'])
missing_data.head(20)
```
